```python
# elimina casualmente il 20% dei valori in una colonna
#1

# Calcola la dimensione della colonna che vuoi modificare
column = 'NoDocbcCost'
column_size = df[column].shape[0]

# Calcola la dimensione dei valori che vuoi eliminare (20% della dimensione della colonna)
drop_size = int(0.0025 * column_size)

# Scegli casualmente i valori da eliminare
drop_index = np.random.choice(df.index, drop_size, replace=False)

# Imposta i valori scelti casualmente a NaN
df.loc[drop_index, column] = np.nan
```
